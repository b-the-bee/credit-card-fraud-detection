Dealing with Imbalanced data using  Under-Sampling

Build a sample dataset containing similar distribtion of normal transcation and fraudulent Transactions.


Number of Fraudulent Transactions--492

```python
legit_sample = legit.sample(n=492)
```

```python
#concatenating Two DataFrames:

new_dataset = pd.concat([legit_sample, fraud], axis =0 )
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

The nature of the data set hasn't changed as the difference between the means from previous values and new values are still there which will help model to detect if the data is legit or fraudulent.

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and Tragets either 0 or 1

```python
x= new_dataset.drop(columns='Class', axis =1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

split the data into tarining data and testing data

```python
#All the features will be stored in X-train 80% of the data
#All the labels will be stored in y-
# 20% of the data will be stored in x-test
# 20% of the corresponding lable will be srored in y-test
#we use stratify to evenly distributing the data
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size = 0.2, stratify = y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
# Training the LogisticsRegresison model with Training data:
model.fit(x_train, y_train)
```

MODEL EVALUATION

Accuracy Score

```python
# accuracy on tarining data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data :', training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print ('accuracy on test data;', test_data_accuracy)
```
